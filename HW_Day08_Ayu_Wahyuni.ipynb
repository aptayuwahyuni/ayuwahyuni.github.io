```python
#Create dictionary
dic = {'JenisGolongan': ['Low','Medium','High','Very High'],
       'min': [0,35.5,70.5,89.9],
       'max': [35.5,70.5,89.9,90]}
#Create dataframe
df_JenisGolongan = pd.DataFrame(data=dic)
#print data
df_JenisGolongan
```
